```python
#Filtering Text- to remove morphological variants or misspelled words in text
def unusual_words(text):
    text_vocab= set(w for w in text if w.isalpha())
    english_vocab= set(w for w in words.words())
    unusual_words=text_vocab.difference(english_vocab)
    return unusual_words
```

```python
unusual_words(gutenberg.words('austen-emma.txt'))
```

```python
#this may be used for the named entity recognition. 
#Find out the names of the males and females ending with each letter of alphabet
#Distribution of the names based on the last charecter of the name among male and female
#first divide the corpus based on condition and then find out the frequency distribution of each group
cfd=nltk.ConditionalFreqDist(
    (fileid, name[-1])
    for fileid in names.fileids()
    for name in names.words(fileid)
)
```

```python
cfd.plot()
```

```python
#Relationship between synsets
#Internally these synsets are stored in form of trie. you can find out the relationship using the tries.
#1. Hyponyms()- is the relationship between subordinates
wordnet.synset('car.n.01').hyponyms()
```

```python
#2 Hypernyms- relationship between superordinates 
wordnet.synset('car.n.01').hypernyms()
```

```python
wordnet.synset('motor_vehicle.n.01').hyponyms()
```

```python
wordnet.synset('motor_vehicle.n.01').hypernyms()
```
